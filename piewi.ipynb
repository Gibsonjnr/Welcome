```python
skills = ["Python", "Javascript", "CSS", "Django"]

# Task 1 — write skills cleanly line by line
with open("my_profile.txt", "w") as file:
    file.write("Name: Ubong Gibson\n")
    file.write("Age: 29\n")
    file.write("Country: Nigeria\n")
    for skill in skills:
        file.write(f"Skill: {skill}\n")   # each skill on its own line

# Task 2 — read file
with open("my_profile.txt", "r") as file:
    print(file.read())

# Task 3 — append
with open("my_profile.txt", "a") as file:
    file.write("Goal: Software Engineer\n")
    file.write("Favourite Language: Python\n")

# Task 4 — read updated file
with open("my_profile.txt", "r") as file:
    print(file.read())

# Task 5 — count lines
with open("my_profile.txt", "r") as file:
    lines = file.readlines()
    print(f"Total lines: {len(lines)}")


# Bonus — fixed version
def save_skills(skills, filename):
    try:
        with open(filename, "w") as file:
            for counter, skill in enumerate(skills, start=1):
                file.write(f"{counter}. {skill}\n")  # "1. Python" format

        # Read file back after writing
        with open(filename, "r") as file:
            print(file.read())

    except Exception as e:
        print(f"Something went wrong: {e}")

skills = ["Python", "Django", "Data Science", "AI Engineering"]
save_skills(skills, "bonus_challenge.txt")
```

```python
def calculate(a, operator, b):
    try :
        a = float(a)
        b = float(b)

        if operator == "+":
            result = a + b
        elif operator == "-":
            result = a - b
        elif operator == "*":
            result = a * b 
        elif operator == "/":

            if b == 0:
                return "Error: cannot divide by zero"
            result = a / b
        else:
            raise ValueError("Invalide oeprator")
        
        return f"{a} {operator} {b} = {result}"
    
    except ValueError as e:
        return f"Error: {e}"
    

print(calculate(10, "+", 5))
```

```python
class BankAccount:
    def __init__(self, owner, balance= 0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return f"Deposited #{amount}. Balance: #{self.balance}"
    
    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient funds!"
        self.balance -= amount
        return f"Withdrew #{amount}. Balance: #{self.balance}"
    
    def __str__(self):
        return f"Account owner: {self.owner} | Balance: #{self.balance}"
    


    


bank = BankAccount("Ubong", 5000)
print(bank.balance)
print(bank.withdraw(200))
print(bank.balance)
print(bank.withdraw(9000))
```

```python
class Developer:

    def __init__(self, name, age, country, skills):
        self.name = name
        self.age = age
        self.country = country
        self.skills = skills

    def introduce(self):
        return f"My name is {self.name}, I am {self.age}, from {self.country}, and my skills are: {self.skills}"
    
    def add_skill(self, skill):
        self.skills.append(skill)
        return f"{skill} added to your skills!"
    

    def remove_skill(self, skill):
        try:
            self.skills.remove(skill)
            return f"You removed {skill} from your skill list!"
        except ValueError:
            return f"{skill} does not exist!!"
        
    def show_skills(self):
        count = 1
        for skill in self.skills:
            print(f"{count} : {skill}")
            count += 1


    def __str__(self):
        return f"Developer: {self.name} | Age: {self.age} | Country: {self.country} | Skills: {', '.join(self.skills)}"

dev = Developer("Ubong Gibson", 29, "Nigeria", ["Python", "Django"])
print(dev.introduce())
print(dev.add_skill("Java"))
print(dev.remove_skill("C"))
print(dev.show_skills())
print(dev)
```

```python
class Developer:

    def __init__(self, name, age, country, skills):
        self.name = name
        self.age = age
        self.country = country
        self.skills = skills

    def introduce(self):
        return f"My name is {self.name}, I am {self.age} years old, from {self.country}."

    def add_skill(self, skill):
        self.skills.append(skill)
        return f"✅ {skill} added to your skills!"

    # ✅ catch ValueError specifically — not bare except
    def remove_skill(self, skill):
        try:
            self.skills.remove(skill)
            return f"✅ {skill} removed from your skills!"
        except ValueError:
            return f"❌ {skill} not found in your skills!"

    # ✅ uses enumerate() and return instead of print
    def show_skills(self):
        result = f"{self.name}'s skills:\n"
        for count, skill in enumerate(self.skills, start=1):
            result += f"  {count}. {skill}\n"
        return result

    def __str__(self):
        return (f"Developer: {self.name} | Age: {self.age} "
                f"| Country: {self.country} | "
                f"Skills: {', '.join(self.skills)}")


# Object 1
dev1 = Developer("Ubong Gibson", 29, "Nigeria", ["Python", "Django"])
print(dev1.introduce())
print(dev1.add_skill("Flask"))
print(dev1.add_skill("Data Science"))
print(dev1.remove_skill("Django"))
print(dev1.remove_skill("Java"))     # skill doesn't exist
print(dev1.show_skills())
print(dev1)

print("---")

# Object 2
dev2 = Developer("Amina Bello", 25, "Ghana", ["Flask", "React"])
print(dev2.introduce())
print(dev2.add_skill("PostgreSQL"))
print(dev2.show_skills())
print(dev2)

```

```python
class Student:
    def __init__(self, name, scores = []):
        self.name = name
        self.scores = scores


    def add_score(self, score):
        self.scores.append(score)
        print(f"Added score {score} for {self.name}. Current scores: {self.scores}")

    def average_score(self):
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)
    
    def grade(self):
        avg = self.average_score()
        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B"
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"
        

    def __str__(self):
        return f"Student: {self.name} | Scores: {self.scores} | Average: {self.average_score():.2f} | Grade: {self.grade()}"
    


work = Student("Ubong", [])
print(work.add_score(86))
print(work.add_score(98))
print(work.add_score(85))

print(work.grade())
    
```

```python
#ENCAPSULATION
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited #{amount}"
        return "Invalid amount"
    
    def withdraw(self, amount):
        if amount > self.__balance:
            return "Insufficient funds!"
        self.__balance -= amount
        return f"Withdrew #{amount}"
    
    def get_balance(self):
        return f"Balance: #{self.__balance}"
    
    def __str__(self):
        return f"Account: {self.owner} | {self.get_balance}"
    

account = BankAccount("Gibson", 50000)
print(account.get_balance())
print(account.withdraw(550))
print(account.get_balance())
print(account.deposit(10000))
print(account.get_balance())


```

```python
class Employee:
    def __init__(self, name, age, salary):
        self.name = name
        self.age = age
        self.salary = salary

    def get_salary(self):
        return self.salary


class Developer(Employee):
    def __init__(self, name, age, salary, skill):
        super().__init__(name, age, salary)
        self.skill = skill


class Designer(Employee):
    def __init__(self, name, age, salary, tool):
        super().__init__(name, age, salary)
        self.tool = tool


class Company:
    def __init__(self, name, employees=None):
        self.name = name
        self.employees = employees if employees is not None else []

    def hire(self, employee):
        self.employees.append(employee)
        return f"Employee {employee.name} added"

    def fire(self, employee):
        self.employees.remove(employee)
        return f"{employee.name} has been fired!"

    def show_team(self):
        return f"{self.name} team: {[employee.name for employee in self.employees]}"

    def total_salary(self):
        total = sum(employee.get_salary() for employee in self.employees)
        return f"Total salary expense: #{total}"

    def __str__(self):
        return f"Company: {self.name} | Employees: {[employee.name for employee in self.employees]}"
    

company = Company("Tech Solutions")

dev = Developer("Ubong", 29, 50000, "Python")
designer = Designer("Amina", 25, 40000, "Figma")

print(company.hire(dev))
print(company.hire(designer))

print(company.show_team())
print(company.total_salary())
```
